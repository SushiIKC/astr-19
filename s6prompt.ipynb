This cell uses a main function to tabulate x with values of sin(x)
and cos(x) in between 2pi with 1000 entries. I used start and end values with a linspace 
command to tabulate the x values, then used np.sin and np.cos to tabulate the two 
trigonometric functions. I then formatted it to appear as a table. 

```python
def main():
    x_values = np.linspace(0,2*math.pi,1000)

    for x in x_values:
        print(f"{x:.4f} | {np.sin(x):.4f} | {np.cos(x):.4f}")
print("{:<8} {:<8} {:<10}".format("x", "sin(x)", "cos(x)"))


if __name__ == "__main__":
    main()
```
